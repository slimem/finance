```
import pandas_datareader as web
import matplotlib.pyplot as plt
import mplfinance as mpf
import datetime as dt
```

```
crypto = "BTC"
currency = "EUR"

start = dt.datetime(2020,1,1)
#end = dt.datetime(2020,2,1)
end = dt.datetime.now()

data = web.DataReader(f"{crypto}-{currency}", "yahoo", start, end)

print(data)
```

```
plt.plot(data['Close'])
plt.show()
```

```
# A candlestick chart for a month of data
smalldata = web.DataReader(f"{crypto}-{currency}", "yahoo", start, dt.datetime(2020,2,1))
mpf.plot(smalldata, type="candle", volume=True, style="yahoo")
```

```
#Compare between BTC and ETH
btc = web.DataReader("BTC-EUR", "yahoo", start, end)
eth = web.DataReader("ETH-EUR", "yahoo", start, end)

plt.yscale("log")

plt.plot(btc['Close'], label="BTC")
plt.plot(eth['Close'], label="ETH")
plt.legend(loc="upper left")
plt.show()
```
